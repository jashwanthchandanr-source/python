```python
def add(n1, n2):
    return n1 + n2

def subtract(n1, n2):
    return n1 - n2

def multiply(n1, n2):
    return n1 * n2

def divide(n1, n2):
    if n2 == 0:
        return "Error! Cannot divide by zero."
    return n1 / n2

operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': divide
}

print("Select operation: +, -, *, /")
operation = input("Enter operation: ")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if operation in operations:
    result = operations[operation](num1, num2)
    print(f"{num1} {operation} {num2} = {result}")
else:
    print("Invalid operation entered")
```

```python
def factorial(n):
    if n < 0:
        return "Error: Factorial is not defined for negative numbers."
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example Usage:
num_for_fact = int(input("Enter a non-negative integer: "))
print(f"The factorial of {num_for_fact} is {factorial(num_for_fact)}.")
```

```python
def product_except_self(nums):
    n = len(nums)
    result = [1] * n

    
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]

    
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]

    return result
```

```python
def isValid(s: str) -> bool:
    stack = []
    # Mapping of closing to opening brackets
    mapping = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in mapping.values():  # opening bracket
            stack.append(char)
        elif char in mapping.keys():  # closing bracket
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()
        else:
            # Ignore non-bracket characters (optional)
            continue
    
    return not stack  # valid if stack is empty


# Test cases
print(isValid("({[]})"))   # True
print(isValid("({[})"))    # False
print(isValid("()[]{}"))   # True
print(isValid("((("))      # False
```

```python
import heapq

def kthLargest(nums, k):
    # Min-heap of size k
    heap = []
    for num in nums:
        heapq.heappush(heap, num)
        if len(heap) > k:
            heapq.heappop(heap)
    return heap[0]

# Example usage
arr = [3, 2, 1, 5, 6, 4]
k = 2
print(kthLargest(arr, k))  # Output: 5
```

```python
def search(nums, target):
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            return mid
        
        # Left half is sorted
        if nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        # Right half is sorted
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    
    return -1  # not found


# Example usage
arr = [4,5,6,7,0,1,2]
target = 0
print(search(arr, target))  # Output: 4
```

```python
from collections import defaultdict

def groupAnagrams(words):
    anagram_map = defaultdict(list)
    
    for word in words:
        # Sort characters to form the key
        key = ''.join(sorted(word))
        anagram_map[key].append(word)
    
    return list(anagram_map.values())


# Example usage
words = ["eat","tea","tan","ate","nat","bat"]
print(groupAnagrams(words))
```

```python
from collections import deque

def maxSlidingWindow(nums, k):
    dq = deque()  # stores indices
    result = []
    
    for i in range(len(nums)):
        # Remove indices out of current window
        if dq and dq[0] <= i - k:
            dq.popleft()
        
        # Remove smaller elements (not useful)
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        
        dq.append(i)
        
        # Append max for current window
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result


# Example usage
arr = [1,3,-1,-3,5,3,6,7]
k = 3
print(maxSlidingWindow(arr, k))  # Output: [3,3,5,5,6,7]
```

```python
def minPlatforms(arrivals, departures):
    n = len(arrivals)
    arrivals.sort()
    departures.sort()
    
    platform_needed = 1
    result = 1
    i, j = 1, 0
    
    while i < n and j < n:
        if arrivals[i] <= departures[j]:
            platform_needed += 1
            i += 1
        else:
            platform_needed -= 1
            j += 1
        
        result = max(result, platform_needed)
    
    return result


# Example usage
arrivals = [900, 940, 950, 1100, 1500, 1800]
departures = [910, 1200, 1120, 1130, 1900, 2000]

print(minPlatforms(arrivals, departures))  # Output: 3
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def hasCycle(head):
    slow, fast = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
```

```python
seats = [0] * 20

while True:
    print("\n1. Book Seat\n2. Cancel Seat\n3. Show Seats\n4. Booking Percentage\n5. Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        seat_no = int(input("Enter seat number (1-20): ")) - 1
        if 0 <= seat_no < 20:
            if seats[seat_no] == 0:
                seats[seat_no] = 1
                print("Seat booked successfully!")
            else:
                print("Seat already booked.")
        else:
            print("Invalid seat number.")

    elif choice == 2:
        seat_no = int(input("Enter seat number (1-20) to cancel: ")) - 1
        if 0 <= seat_no < 20:
            if seats[seat_no] == 1:
                seats[seat_no] = 0
                print("Reservation cancelled.")
            else:
                print("Seat was not booked.")
        else:
            print("Invalid seat number.")

    elif choice == 3:
        print("Seats (0=Available, 1=Booked):", seats)

    elif choice == 4:
        booked_count = seats.count(1)
        percentage = (booked_count / 20) * 100
        print(f"Booking Percentage: {percentage}%")

    elif choice == 5:
        break
```
